```python
# lambda function to extract first number in a string
get_nums = lambda x: [float(num) for num in str(x).split() if num.isdigit()]
get_first_num_else_nan = lambda x: x[0] if len(x) > 0 else np.nan
```

```python
# Problem row, change to just 7
# data[data['What is your favorite number?'] == '7 and 2']['What is your favorite number?'] = '7'
data['What is your favorite number?'].replace('7 and 2', '7', inplace=True)
data['What is your favorite number?'].replace('Don\'t have one.', np.nan, inplace=True)

# Change to floats
data['What is your favorite number?'] = data['What is your favorite number?'].astype(float)
```

```python
# Change to lists of floats and then get first element if it exists
data['What temperature is it outside right now? (In Celsius) '] = data['What temperature is it outside right now? (In Celsius) '].apply(lambda x: [float(num) for num in str(x).split() if num.isdigit()])
data['What temperature is it outside right now? (In Celsius) '] = data['What temperature is it outside right now? (In Celsius) '].apply(lambda x: x[0] if len(x) > 0 else np.nan)
```
